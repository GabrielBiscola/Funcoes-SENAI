```python
def calculadora(): #Calculadora
  options=[1,2,3,4,5,6,7]
  exibir_menu()
  option=int(input())
  if option not in options:
    print('Calculadora encerrada!')
    return (False)

  else:
    if option == 1:
      print('O cálculo será feito da seguinte maneira: n1 + n2 (n1 mais n2)')
      a, b = ler(option)
      print(f'Resultado: {soma(a, b)}')
    elif option == 2:
      print('O cálculo será feito da seguinte maneira: n1 - n2 (n1 menos n2)')
      a, b = ler(option)
      print(f'Resultado: {sub(a, b)}')
    elif option == 3:
      print('O cálculo será feito da seguinte maneira: n1 x n2 (n1 vezes n2)')
      a, b = ler(option)
      print(f'Resultado: {mult(a, b)}')
    elif option == 4:
      print('O cálculo será feito da seguinte maneira: n1 / n2 (n1 dividido por n2)')
      a, b = ler(option)
      print(f'Resultado: {div(a, b)}')
    elif option == 5:
      print('O cálculo será feito da seguinte maneira: n1^n2 (n1 = base, n2 = expoente)')
      a, b = ler(option)
      print(f'Resultado: {pot(a, b)}')
    elif option == 6:
      print('O cálculo será feito da seguinte maneira: Radicando = n1 Índice = n2\nExemplo: Raiz quadrada de 9 = Radicando 9 índice 2 = 2V9')
      a, b = ler(option)
      print(f'Resultado: {raiz(a, b)}')
    elif option == 7:
      print('O cálculo será feito da seguinte maneira: n! = n*(n-1)*(n-2)...')
      a = ler(option)
      print(f'Resultado: {fat(a)}')

def exibir_menu():
  print('Escolha a operação:\n 1-Soma\n 2-Subtração\n 3-Multiplicação\n 4-Divisão\n 5-Potenciação\n 6-Radiciação\n 7-Fatoração')


def ler(option):
  if option < 7:
    n1 = float(input('N1:'))
    n2 = float(input('N2:'))
    return (n1, n2)
  else:
    n1 = float(input('N1:'))
    return (n1)

def soma(a, b):
  return (a+b)
def sub(a, b):
  return (a-b)
def mult(a, b):
  return (a*b)
def div(a, b):
  if b != 0:
    return (a/b)
  else:
    print('Atenção: Você está tentando dividir um número por 0.')
def pot(a, b):
  return (a**b)
def raiz(a, b):
  if a >= 0 or b % 2 != 0:
    return (a**(1/b))
  else:
    print('Atenção: Você está tentando extair uma raiz par de um número negativo.')
def fat(a):
  fat = 1
  if a >= 0:
    if a > 0:
      for i in range (1, a+1):
        fat *= i
    return (fat)
  else:
    print('Atenção: Você está tentando calcular o fatorial de um número negativo.')

while True:
  if calculadora() == False:
    break
```

```python
#Verificador de CPF
def validar_digito_verificador_1(digitos):
  soma_prod = 0
  for i in range(9):
    soma_prod += digitos[i]*(10-i)
  d1 = soma_prod%11
  if d1 < 2:
    d1 = 0
  else:
    d1 = 11-d1
  if d1 == digitos[-2]:
    return True
  else:
    return False


def validar_digito_verificador_2(digitos):
  soma_prod = 0
  for i in range(10):
    soma_prod += digitos[i]*(11-i)
  d2 = soma_prod%11
  if d2 < 2:
    d2 = 0
  else:
    d2 = 11-d2
  if d2 == digitos[-1]:
    return True
  else:
    return False

def validar_cpf(cpf):
  print(f'CPF = {cpf}')
  cpfn = [int(i) for i in cpf]
  print(f'CPFN = {cpfn}')
  v1 = validar_digito_verificador_1(cpfn)
  if v1 == False:
    return False
  else:
    v2 = validar_digito_verificador_2(cpfn)
    if v2:
      return True
    else:
      return False

def main():
  cpf = input('Digite o cpf (apenas números): ')
  if validar_cpf(cpf) == True:
    print('CPF válido!')
  else:
    print('CPF inválido!')

main()
```

```python
#Gerar CPF
def gerar_digito_verificador_1(digitos):
  soma_prod = 0
  for i in range(9):
    soma_prod += digitos[i]*(10-i)
  d1 = soma_prod%11
  if d1 < 2:
    d1 = 0
  else:
    d1 = 11-d1
  digitos.append(d1)
  return digitos

def gerar_digito_verificador_2(digitos):
  soma_prod = 0
  for i in range(10):
    soma_prod += digitos[i]*(11-i)
  d2 = soma_prod%11
  if d2 < 2:
    d2 = 0
  else:
    d2 = 11-d2
  digitos.append(d2)
  return digitos

def gerar_cpf(cpf9):
  cpfn9 = [int (i) for i in cpf9]
  cpfn10 = gerar_digito_verificador_1(cpfn9)
  cpfn = gerar_digito_verificador_2(cpfn10)
  cpfstr = "".join(str(x) for x in cpfn)
  return (cpfstr[:9])+"-"+(cpfstr[9:])

def main():
  cpf9 = input('Digite os primeiros 9 dígitos do CPF a ser gerado (apenas números):')
  cpf = gerar_cpf(cpf9)
  print(f'CPF = {cpf}')

main()
```
